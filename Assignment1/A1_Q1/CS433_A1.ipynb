```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
times={}
times[20] = [622361, 331834, 182209, 123202]
times[21] = [1407986, 748124, 405789, 270962]
times[22] = [3140183, 1684550, 904696, 605850]
times[23] = [6983380, 3725164, 2022612, 1335585]
times[24] = [15515952, 8295695, 4424397, 2956555]

speedups = {}
for i in range(5):
    speedups[i + 20] = 1 / (np.array(times[i + 20]) / times[i + 20][0])
```

```python
x = [1,2,4,8]

colors = ['red', 'blue', 'orange', 'green', 'purple']

plt.xlabel('Number of threads')
plt.ylabel('Time (in ms)')

for i in range(5):
    plt.plot(x, times[20 + i], 'ks-', color = colors[i])

plt.legend(['N = 20', 'N = 21', 'N = 22', 'N = 23', 'N = 24'], loc = 'upper right')
plt.savefig('Time-vs-threads.png')
plt.show()
```

```python
x = [1,2,4,8]

colors = ['red', 'blue', 'orange', 'green', 'purple']

plt.xlabel('Number of threads')
plt.ylabel('Speedup')

for i in range(5):
    plt.plot(x, speedups[20 + i], 'ks-', color = colors[i])

# plt.axhline(y = 1, color = 'black', linestyle = '--')
# plt.axhline(y = 2, color = 'black', linestyle = '--')
# plt.axhline(y = 4, color = 'black', linestyle = '--')

plt.legend(['N = 20', 'N = 21', 'N = 22', 'N = 23', 'N = 24'], loc = 'lower right')
plt.savefig('Speedup-vs-threads.png')
plt.show()
```
